```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
print(tips.to_string())
```

```python
tips["time"].unique()
# tips["day"].unique()
```

```python
sns.set_theme(style="whitegrid")
sns.displot(tips['total_bill'], kde = True)
```

Displot Fonksiyonu

sns.displot(tips['total_bill'], kde = True)
sns.displot(data = tips, x= 'total_bill', kde = True)

Her iki kod da aynı grafiği üretir, sadece parametrelerin kullanım şekli farklıdır. İkinci yaklaşım, özellikle birden fazla değişkenle çalışırken veya daha karmaşık grafikler oluştururken daha esnek ve okunabilir olabilir.

(kde: kernel density estimation)

```python
sns.set_theme(style='ticks')
sns.displot(data = tips, x= 'total_bill', bins=28, kde = True)
```

```python
sns.displot(tips['tip'], kind = "kde")
# sns.kdeplot(tips['tip']) iki kod aynı sonucu verir
```

```python
sns.kdeplot(tips['total_bill'])
```

seaborn'daki displot fonksiyonu, temel olarak tek değişkenli (univariate) dağılımları (histogram, KDE, vb.) çizmek için kullanılır. Bu nedenle, displot ile x ve y eksenine farklı değişkenler ekleyemezsin. displot, yalnızca tek bir değişkenin dağılımını görselleştirmek için tasarlanmıştır.

Ancak, iki değişken arasındaki ilişkiyi görselleştirmek istiyorsan, başka bir seaborn fonksiyonu olan scatterplot, jointplot, veya relplot gibi fonksiyonları kullanabilirsin. Özellikle jointplot, hem iki değişkenin dağılımını hem de aralarındaki ilişkiyi göstermek için idealdir.

```python
sns.jointplot(data=tips, x='time', y='tip')
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip')
```

```python
sns.jointplot(data=tips, x='total_bill', y='size')
```

Farklı veri çerçeveleriyle çalışırken x'i bir veri çerçevesi, y'yi başla bir veri çerçevesinden alabilirsin.
Örn:

```python
sns.jointplot(x=tips['total_bill'], y=tips['tip'], kind='hex')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde')
```

```python
sns.pairplot(tips)
```
