```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = [1, 2, 3, 4, 5]  # Bu, x değişkenindeki veriyi tutar
y = [2, 4, 6, 8, 10]  # Bu da y değişkenindeki veriyi tutar
plt.subplot(1, 2, 1)  # 1 satır, 2 sütun, 1. grafik
plt.plot(x, y, 'r')  # 'x' yatayda, 'y' dikeyde
plt.subplot(1, 2, 2)  # 1 satır, 2 sütun, 2. grafik
plt.plot(y, x, 'b')  # 'y' yatayda, 'x' dikeyde

```

```python
fig, axes = plt.subplots(1, 3)  

# Her eksene ayrı grafik ekleyelim
axes[0].plot([1, 2, 3], [4, 5, 6])  # 1. subplot
axes[1].plot([1, 2, 3], [6, 5, 4])  # 2. subplot
axes[2].plot([1, 2, 3], [5, 5, 5])  # 3. subplot

# Başlık ekleyelim
axes[0].set_title("İlk Grafik")
axes[1].set_title("İkinci Grafik")
axes[2].set_title("Üçüncü Grafik")
plt.tight_layout()
# Grafik göster
plt.show()
```

```python
fig = plt.figure(figsize=(5, 5))  # Büyük bir figür oluştur
axes = fig.add_axes([0.1, 0.1, 0.4, 0.4])  # Küçük bir eksen ekle
axes.plot([1, 2, 3], [4, 5, 6])
plt.show()
```

```python
fig = plt.figure(figsize=(5, 5))
axes = fig.add_axes([0, 0, 1, 1])  # Figürün tamamını kaplayan eksen
axes.plot([1, 2, 3], [4, 5, 6])
plt.show()
```
