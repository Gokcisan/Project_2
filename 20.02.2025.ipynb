```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = np.linspace(0, 5, 11)
y = x**2
x, y
```

```python
plt.plot(y)
plt.xlabel('X Label')
plt.ylabel("Y Label")
plt.title('A Very Descriptive Title')
```

```python
plt.subplot(1,3,1)
plt.subplot(1,3,2)
plt.plot(y, x)
plt.subplot(1,3,3)
plt.tight_layout()
plt.plot(x, y)
```

tek bir figür (grafik alanı) içinde yan yana 3 ayrı alt grafik (subplot) oluştururken kod bu şekilde yazılıyorsa subplot(x, y, z) değerlerinden bazılarının hep aynı olması gerekir, yoksa hata verir değil mi?

Evet, doğru düşünüyorsun! plt.subplot(x, y, z) fonksiyonunda bazı değerlerin tutarlı olması gerekir, yoksa hata alırsın.

Parametrelerin Anlamı:
x → Satır sayısı (kaç satır olacak?)
y → Sütun sayısı (kaç sütun olacak?)
z → Hangi alt grafiği seçtiğimiz (1'den başlayarak numaralandırılır)
Hata Alınabilecek Durumlar
Eğer x ve y değerleri her subplot için farklı olursa, Python hata vermez, ama beklenmedik bir şekilde grafikleri çizer ve düzen bozulur.

Ancak z değeri geçerli bir sayı olmazsa, hata alırsın.

plt.subplot(1, 3, 4)  # HATA! Çünkü 1 satırda 3 sütun var ama 4. grafik diye bir şey olamaz.

Doğru Kullanımın Mantığı
Eğer 3 tane yan yana grafik istiyorsan:

x (satır) aynı olmalı → 1
y (sütun) aynı olmalı → 3
z (pozisyon) 1'den 3'e kadar olmalı → 1, 2, 3

Özetle, ilk iki parametre (x, y) her subplot çağrısında aynı olmalı, z ise sırasıyla 1'den başlayarak artmalı.

yani bu yazım "plt.subplot(1,3,1)
plt.subplot(1,3,2)
plt.subplot(1,3,3)" aslında bir matris için boş bir şablon oluşturuyor diyebilir miyiz?
Evet, çok doğru bir bakış açısı! 👌

Bu yazım biçimi 1x3 boyutunda bir matris için boş bir şablon oluşturuyor diyebiliriz.
Ama grafikler henüz çizilmediği için boş bir düzen (şablon) oluşur. İçine plt.plot(), plt.scatter(), plt.bar() gibi grafik komutları ekleyerek doldurabilirsin.

Bu mantıkla subplot kullanımı aslında matris gibi düşünülebilir. 🎯

```python
plt.figure(figsize=(9,9))

plt.subplot(2, 2, 1)  # 1. grafik (1. satır, 1. sütun)
plt.title("Grafik 1")

plt.subplot(2, 2, 2)  # 2. grafik (1. satır, 2. sütun)
plt.title("Grafik 2")

plt.subplot(2, 2, 3)  # 3. grafik (2. satır, 1. sütun)
plt.title("Grafik 3")

plt.subplot(2, 2, 4)  # 4. grafik (2. satır, 2. sütun)
plt.title("Grafik 4")

plt.show()
```

```python
plt.subplot(2,3,1)
plt.subplot(2,3,2)
plt.subplot(2,3,3)
plt.subplot(2,3,4)
plt.subplot(2,3,5)
plt.subplot(2,3,6)
plt.tight_layout()
```

```python
plt.subplot(1, 2, 1)
plt.title("Grafik 1")
plt.plot(x, y, 'r')
plt.xlabel("X Label")
plt.ylabel('Y Label')

plt.subplot(1, 2, 2)
plt.plot(y, x, 'b')
plt.title("Grafik 2")
plt.ylabel('X Label')
plt.xlabel('Y Label')
plt.tight_layout()
```

```python
# OOM Object Oriented Method

fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x, y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('X = Y^2 Graph')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.4, 0.3, 0.3])

axes1.plot(x, y)
axes1.set_xlabel('X Axis')
axes1.set_title('Big Plot')
axes2.plot(y, x)
axes2.set_ylabel('X Axis')
axes2.set_title('Small Plot')
```

```python
fig, axes = plt.subplots(1, 3)  # 1 satır, 3 sütun subplot

axes[0].set_title("Birinci subplot")
axes[1].set_title("İkinci subplot")
axes[2].set_title("Üçüncü subplot")
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(nrows=3, ncols=3)
plt.tight_layout()
```

```python
fig, axes = plt.subplots(nrows= 1, ncols=3)

for i in axes:
    i.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows= 1, ncols=2)

axes[0].plot(x, y, 'r')
axes[0].set_title('1st Plot')
axes[1].plot(y, x, 'b')
axes[1].set_title("2nd Plot")
```
